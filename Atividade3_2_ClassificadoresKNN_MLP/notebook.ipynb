### Conjunto de dados

#### Leitura do conjunto de dados

```python
import pandas as pd

data = pd.read_csv("./data/data1.txt" , sep=",", header=None)
data.head()
```

```python
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values.reshape(-1,1)
```

#### Normalizando os dados

```python
from Utils import normalization
```

```python
X, _ = normalization(X, X)
```

### Relatório

#### KNN

```python
# k = 1
knn = KNN( 1 )
acc = kfold( X, Y, 5, knn )
print( "KNN com k=1 => acurácia de", acc )
```

```python
knn.fit( X, Y )
plot_boundaries( X , Y, knn)
```

```python
# k = 2
knn = KNN( 2 )
acc = kfold( X, Y, 5, knn )
print( "KNN com k=2 => acurácia de", acc )
```

```python
knn.fit( X, Y )
plot_boundaries( X , Y, knn)
```

```python
# k = 3
knn = KNN( 3 )
acc = kfold( X, Y, 5, knn )
print( "KNN com k=3 => acurácia de", acc )
```

```python
knn.fit( X, Y )
plot_boundaries( X , Y, knn)
```

#### MLP

```python
# hidden layer size = 2
mlp = MultiLayerPerceptron( hidden_layer_size=2 )
acc = kfold( X, Y, 5, mlp )
print( "MLP com 2 neurônios na camada oculta => acurácia de", acc )
```

```python
mlp.fit( X, Y )
plot_boundaries( X , Y, mlp)
```

```python
# hidden layer size = 3
mlp = MultiLayerPerceptron( hidden_layer_size=3 )
acc = kfold( X, Y, 5, mlp )
print( "MLP com 3 neurônios na camada oculta => acurácia de", acc )
```

```python
mlp.fit( X, Y )
plot_boundaries( X , Y, mlp)
```

```python
# hidden layer size = 2
mlp = MultiLayerPerceptron( hidden_layer_size=4 )
acc = kfold( X, Y, 5, mlp )
print( "MLP com 4 neurônios na camada oculta => acurácia de", acc )
```

```python
mlp.fit( X, Y )
plot_boundaries( X , Y, mlp)
```
